```python
X_tr, X_ts, y_tr, y_ts = train_test_split(df.iloc[:,1:],df.iloc[:,0], test_size=0.2)
```

```python
scaler = StandardScaler()
X_tr = scaler.fit_transform(X_tr)
X_ts = scaler.transform(X_ts)
```

```python
enc = LabelEncoder()
y_tr = enc.fit_transform(y_tr)
y_ts = enc.transform(y_ts)
```

**Using Tensors**

```python
X_tr_tensor = torch.from_numpy(X_tr)
X_ts_tensor = torch.from_numpy(X_ts) 
y_tr_tensor = torch.from_numpy(y_tr)
y_ts_tensor = torch.from_numpy(y_ts) 
```

**Model Initialization**

```python
class SimpleNeuralNet():

    def __init__(self, X):
        self.weights = torch.rand(X.shape[1],1,dtype=torch.float64,requires_grad=True)
        self.bias  = torch.zeros(1,dtype=torch.float64, requires_grad=True)

    def forward(self,X):
        z = torch.matmul(X, self.weights) + self.bias
        ypred = torch.sigmoid(z)
        return ypred
    
    def loss_func(self, ypred, y):
        epsilon = 1e-7
        ypred = torch.clamp(ypred, epsilon, 1 - epsilon)

        loss = -(y_tr_tensor * torch.log(ypred) + (1-y_tr_tensor) * torch.log(1 - ypred)).mean()
        return loss
```

```python
model = SimpleNeuralNet(X_tr_tensor)

learning_rate = 0.1
epochs = 50
for epoch in range(epochs):
    ypred = model.forward(X_tr_tensor)

    loss = model.loss_func(ypred, y_tr_tensor)

    loss.backward()

    with torch.no_grad():
        model.weights -= learning_rate * model.weights.grad
        model.bias -= learning_rate * model.bias.grad

    model.weights.grad.zero_()
    model.bias.grad.zero_()

    print(f"Epoch: {epoch + 1},  Loss: {round(loss.item(), 6)}")
```

```python
with torch.no_grad():
    ypred = model.forward(X_ts_tensor)
    ypred = (ypred > 0.9).float()
    acc = (ypred == y_ts_tensor).float().mean()
    print(f"Accuracy: {acc}")
```
